```python
import numpy as np
import rebound
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution, root, least_squares
from scipy.spatial.transform import Rotation as R
from scipy.interpolate import CubicSpline
import os, sys
import gc
import pandas as pd

from importlib import reload
import integrator
reload(integrator)
from integrator import *
```

```python
default_configs = {
    'planet_num': 2,
    'planet_mass': [1e-7, 1e-7],
    'kappa': 2,
    'inner_period': 2*np.pi,
    'C': [],
    'target_mean_anomaly': 4*np.pi,
    'init_time_step': 0.2,
    'bisection_tol': 1e-11,
    'bisection_doom': 50000,
}
```

```python
# Mass Continuation then Kappa

mass_cands = (10 ** np.arange(-9., -4., 1.))
kappa_cands = np.linspace(1.999, 2.001, 31)

bounds = [(-3, -0.5), (-3, -0.5), (-np.pi, np.pi), (-np.pi, np.pi)]

# with open('figures/configs_list.txt', 'w') as f:
#         f.write(f'index, res, mse, configs\n')

results = pd.DataFrame(columns=['kappa', 'res', 'res_fun', 'mse', 'configs'])

for j, kappa in enumerate(kappa_cands):
    init_theta = [-1, -1, 0, 0]
    
    for i, m in enumerate(mass_cands):
        m_configs = default_configs.copy()
        m_configs['planet_mass'] = [m, m]
        m_configs['kappa'] = kappa
    
        res_0 = minimize(optimizing_function, x0=init_theta, method='Nelder-Mead', bounds=bounds, args=(m_configs,), options={ 'xatol':1e-12, 'fatol':1e-12})
        # res_0 = minimize(optimizing_function, x0=init_theta, method='Nelder-Mead', bounds=bounds, args=(m_configs,), tol=1e-16)
        gc.collect()
        print(f'Masses {m}, kappa {kappa}, first done!')
        res = minimize(optimizing_function, x0=res_0.x, method='L-BFGS-B', args=(m_configs,), bounds=bounds, tol=1e-16)
        print(f'Masses {m}, kappa {kappa}, second done!')
        gc.collect()
        init_theta = res.x
        print(res.x, res.fun)
    
    mse = calculate_mse(res.x, m_configs)
    print(res.x, res.fun, mse)

    marks = np.arange(0, 5001, 1)

    sim = init_simulation(res.x, m_configs)
    _, target_time, _, _ = integrate_one_cycle(sim, m_configs)
    print(m_configs)
    
    angle_hist = np.zeros(len(marks))
    pomega_1_hist = np.zeros(len(marks))
    pomega_2_hist = np.zeros(len(marks))
    
    for i, _ in enumerate(marks):
        sim.integrate(sim.t + target_time)
        sim.t = 0
        sim.move_to_hel()
        
        angle_hist[i] = sim.particles[1].l - 2*sim.particles[2].l + sim.particles[1].pomega
    
        pomega_1_hist[i] = sim.particles[1].pomega
        pomega_2_hist[i] = sim.particles[2].pomega
    
        # print(sim.particles[1].l, sim.particles[2].l, sim.particles[1].pomega)

    plt.subplots(2, 1, figsize=(6,8))
    plt.tight_layout()
    plt.subplot(2, 1, 1)
    plt.plot(wrap_angles(angle_hist))
    plt.title(f'resonant argument (kappa = {m_configs['kappa']})')
    
    plt.subplot(2, 1, 2)
    plt.plot(wrap_angles(pomega_1_hist), label='1')
    plt.plot(wrap_angles(pomega_2_hist), label='2')
    plt.plot(wrap_angles(pomega_1_hist - pomega_2_hist), label='1 - 2')
    plt.legend()
    plt.title(f'pomega (kappa = {m_configs['kappa']})')

    results.loc[len(results)] = [kappa, res.x, res.fun, mse, m_configs]

    # with open('figures/configs_list.txt', 'a') as f:
    #     f.write(f'{str(j).zfill(4)}, {res.x}, {mse}, {m_configs}\n')
    plt.savefig(f'figures/plot_{str(j).zfill(4)}')
    results.to_pickle('figures/data.pkl')
    
    plt.show()
        
```

```python
from sklearn.model_selection import ParameterGrid

mass_grid = {'mass_1': np.arange(1e-4, 101e-4, 10e-4),
             'mass_2': np.arange(1e-4, 101e-4, 10e-4),}

# WITH Mass continuation (4 times logarithmically)

bounds = [(-3, -0.5), (-3, -0.5), (-np.pi, np.pi), (-np.pi, np.pi)]

T_vs_mass_contd_results = pd.DataFrame(columns=['mass_1', 'mass_2', 'T', 'res', 'res_fun', 'mse', 'configs'])

init_theta = [-1, -1, 0, 0]
kappa = 2.

for masses in ParameterGrid(mass_grid):
    init_theta = [-1, -1, 0, 0]

    submass_1 = masses['mass_1'] * (10 ** np.arange(-6., 1., 2))
    submass_2 = masses['mass_2'] * (10 ** np.arange(-6., 1., 2))

    for sm1, sm2 in zip(submass_1, submass_2):
        m_configs = default_configs.copy()
        m_configs['planet_mass'] = [sm1, sm2]
        m_configs['kappa'] = kappa
    
        res_0 = minimize(optimizing_function, x0=init_theta, method='Nelder-Mead', bounds=bounds, args=(m_configs,), options={ 'xatol':1e-12, 'fatol':1e-12})
        # res_0 = minimize(optimizing_function, x0=init_theta, method='Nelder-Mead', bounds=bounds, args=(m_configs,), tol=1e-16)
        gc.collect()
        print(f'Sub-Masses {sm1, sm2}, kappa {kappa}, first done!')
        res = minimize(optimizing_function, x0=res_0.x, method='L-BFGS-B', args=(m_configs,), bounds=bounds, tol=1e-16)
        print(f'Sub-Masses {sm1, sm2}, kappa {kappa}, second done!')
        gc.collect()
        init_theta = res.x
    print(res.x, res.fun)
    
    mse = calculate_mse(res.x, m_configs)
    print(res.x, res.fun, mse)
    
    sim = init_simulation(res.x, m_configs)
    _, target_time, _, _ = integrate_one_cycle(sim, m_configs)
    print(target_time)
    print(m_configs)
    print()

    T_vs_mass_contd_results.loc[len(T_vs_mass_contd_results)] = [masses['mass_1'], masses['mass_2'], target_time, res.x, res.fun, mse, m_configs]
    # with open('figures/configs_list.txt', 'a') as f:
    #     f.write(f'{str(j).zfill(4)}, {res.x}, {mse}, {m_configs}\n')
    # plt.savefig(f'figures/plot_{str(j).zfill(4)}')
    # results.to_pickle('figures/data.pkl')
    
    plt.show()
    
```
